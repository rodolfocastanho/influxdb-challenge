```python
# Trata Site_Apoio
def fill_SiteApoio(row):
    siteAApoio = row['SITE_A_APOIO']
    siteBApoio = row['SITE_B_APOIO']
    for i in arraySitesDePara:
        if siteAApoio == i[4]:
            siteAApoio = i[5]
        if siteBApoio == i[4]:
            siteBApoio = i[5]    
    return siteAApoio, siteBApoio            
```

```python
# Trata Int_Apoio
def fill_IntApoio(row):
    intAApoio = row['INT_A']
    intBApoio = row['INT_B']
    bundle = row['Bundle']
    
    if str(row['INT_A']).lower().startswith(('eth-trunk','bundle-ether','port-channel', 'ae')) or str(row['INT_B']).lower().startswith(('eth-trunk','bundle-ether','port-channel', 'ae')):
        if '.' not in row['INT_A'].lower() and '.' not in row['INT_B'].lower():
            bundle = 'B'
        else:
            bundle = 'L'
    
    if row['HOST_B'] in row['INT_A']:
        intAApoio = row['MEMBROS_AGREGADA_A']
    if row['HOST_A'] in row['INT_B']:
        intBApoio = row['MEMBROS_AGREGADA_B']
            
    return intAApoio, intBApoio, bundle        
```

```python
# Campo: SP_FSP
# Informação: UF_A / UF_B
# Passos: 
    # 1 - Se ambos = SP então SP senão FSP
def fill_SP_FSP(row):
    return 'SP' if row['UF_A'] == 'SP' and row['UF_B'] == 'SP' else 'FSP'
```

```python
# Campo: Velocidade
# Informação: BW_A / BW_B
# Passos: 
    # 1 - Substituir N/A (string) para 0 (zero)
    # 2 - Comparar os campos e trazer o maior
    # 3 - Dividir o valor por 1 Bilhão (1x10^9)
    # 4 - Se = 1,024 então 1
def fill_Velocidade(row):
    bwA = 0 if str(row['BW_A']) == 'nan' else row['BW_A']
    bwB = 0 if str(row['BW_B']) == 'nan' else row['BW_B']
    
    rowValue = bwA if bwA > bwB else bwB
    rowValue = rowValue / (10**9)
    return 1 if rowValue == 1.024 else rowValue
```

```python
# Campo: TesteVelocidade
# Informação: Velocidade
# Passos: 
    # 1 - Comparação:
        #Se = 1 então GBE
        #Se = 10 então GBE10
        #Se = 100 então GBE100
def fill_TesteVelocidade(row):
    rowValue = row['Velocidade']
    # Alterar para <=1 G ou M
    if rowValue >= 1:
        if rowValue == 1:
            return 'GBE'
        else:
            return 'GBE'+str(round(rowValue))
    elif rowValue == 0:
        return ''
    else:
        return str(round(rowValue*1000))+'M'


def fill_TesteVelocidade2(row):
    rowValue = row['Velocidade']
    if rowValue == 1:
        return 'GBE'
    elif rowValue == 10:
        return 'GBE10'
    elif rowValue == 100:
        return 'GBE100'
    else:
        return ''
```

```python
# Campo: EILD
# Informação: DESCR_A / DESCR_B
# Tipo: 
    # Se SP_FSP = SP então EILD = NULO
    # 8 caracteres formato 99000000
# Passos: 
    # 1 - Buscar 10 caracteres após 99
    # 2 - Retirar o que não é número
    # 3 - Se 10 caracteres então é ODU (ignorar)
    
# TODO: verificar eild duplicado
    # dois na DESCR_A ou DESCR_B
    # se um em cada, comparar se é o mesmo
    
def trataEILD(descr, array):
    eild = re.search("99[0-9]{8}|99[0-9]{6}", descr)
    
    if eild == None:
        if len(array)==1 and len(array[0])==8:
            eild = array[0]
        else:
            eild = ''
        return eild, array
    else:
        array.append(eild[0])
        return trataEILD(descr.replace(str(eild[0]), ''), array)

def fill_EILD(descrA, descrB, array):
    eildA, arrayA = trataEILD(descrA, [])
    eildB, arrayB = trataEILD(descrB, [])
    if len(eildA)>0 and ((len(eildB)>0 and eildA==eildB) or len(eildB)==0):
        return eildA, ''
    elif len(eildB)>0 and ((len(eildA)>0 and eildA==eildB) or len(eildA)==0):
        return eildB, ''
    else:
        return '', str(set(arrayA+arrayB)).replace('{','').replace('}','').replace('set()','')
              
```

```python
# Campo: Designador
# Informação: 
# Tipo: 
    # Prioridade: Se encontrar EILD
        # senão busca GBE
# Passos: 
    
def defineMeioTx(siteA, siteB, descrA, descrB, velocidade, fsp):
    meioTxA = ''
    meioTxB = ''
    checkEILD = False
    checkGBE = False
    eild, EILD_Dupl = fill_EILD(descrA, descrB, [])
    
    if len(eild)>0:
        meioTxA, meioTxB, checkEILD = 'EILD'+str(eild), 'EILD'+str(eild), True
    elif siteA==siteB:
        meioTxA, meioTxB = 'LOCAL', 'LOCAL'
    else:
        meioTxA, meioTxB, checkGBE = defineMeioTxGBE(descrA, descrB, velocidade)
        if not checkGBE:
            meioTxA, meioTxB, checkGBE = defineMeioTxFibra(descrA, descrB)
    
    return meioTxA, meioTxB, checkEILD, checkGBE, EILD_Dupl

def defineMeioTxGBE(descrA_full, descrB_full, velocidade):
    padraoA = ''
    padraoB = ''
    meioTxA = ''
    meioTxB = ''
    checkGBE = False
    descrA = re.search(str(velocidade), descrA_full)
    descrB = re.search(str(velocidade), descrB_full)
    
    if descrA != None:
        padraoA = descrA_full[descrA.start()-16:descrA.end()+5]
    if descrB != None:
        padraoB = descrB_full[descrB.start()-16:descrB.end()+5]
    splitA = padraoA.split('-')
    splitB = padraoB.split('-')
    
    if len(padraoA)>0 and len(padraoB)>0:
        if padraoA == padraoB:
            meioTxA = padraoA if len(splitA)==4 and len(splitA[0])==7 and len(splitA[1])==7 and splitA[2]==velocidade else ''
            meioTxB = padraoB if len(splitB)==4 and len(splitB[0])==7 and len(splitB[1])==7 and splitB[2]==velocidade else ''
    
    if len(meioTxA)>0 or len(meioTxB)>0:
        checkGBE = True
        
    return meioTxA, meioTxB, checkGBE

def defineMeioTxFibra(descrA_full, descrB_full):
    padraoA = ''
    padraoB = ''
    meioTxA = ''
    meioTxB = ''
    checkGBE = False
    descrA = re.search('-SO-', descrA_full)
    descrB = re.search('-SO-', descrB_full)
    
    if descrA != None:
        padraoA = descrA_full[descrA.start()-15:descrA.end()+4]
    if descrB != None:
        padraoB = descrB_full[descrB.start()-15:descrB.end()+4]
    splitA = padraoA.split('-')
    splitB = padraoB.split('-')
    
    if len(padraoA)>0 and len(padraoB)>0:
        if padraoA == padraoB:
            meioTxA = padraoA if len(splitA)==4 and len(splitA[0])==7 and len(splitA[1])==7 and splitA[2]=='SO' else ''
            meioTxB = padraoB if len(splitB)==4 and len(splitB[0])==7 and len(splitB[1])==7 and splitB[2]=='SO' else ''
    
    if len(meioTxA)>0 or len(meioTxB)>0:
        checkGBE = True
        
    return meioTxA, meioTxB, checkGBE

def fill_Designador(row):
    designadorA, designadorB, checkEILD, checkGBE, EILD_Dupl = defineMeioTx(row['SITE_A_APOIO'], 
                                                                 row['SITE_B_APOIO'], 
                                                                 row['DESCR_A'], row['DESCR_B'], 
                                                                 row['TesteVelocidade'], row['SP_FSP'])
    designador =  designadorA if len(designadorA)>0 else designadorB
    designadorTX = True if len(designador)>0 and designador!='LOCAL' else False
    #designadorTX = True if len(designador)>0 else False
    return designador, checkEILD, checkGBE, designadorTX, EILD_Dupl
```

```python
# Designadores não encontrados. Verificar se é Vivo2
def fill_EildVivoII(descrA, descrB, df):
    buscaA = df['EILD'][(df['SiglaRota']==descrA.split(':')[-1].upper())]
    buscaB = df['EILD'][(df['SiglaRota']==descrB.split(':')[-1].upper())]
    
    if len(buscaA>0):
        eild = 'EILD'+str(buscaA.values[0])
        designador = True
    elif len(buscaB>0):
        eild = 'EILD'+str(buscaB.values[0])
        designador = True
    else:
        eild = ''
        designador = False
        
    return eild, designador
```

```python
# Tratamento de Interface (Campos APOIO)

# Substituir textos
def interfacesPrefixo(row):
    def abrevia(interface):
        for i in arrayInterface:        
            prefixoCompleto = i[0]
            prefixoAbreviado = i[1]

            if interface[0:len(prefixoCompleto)] == prefixoCompleto:
                interface = interface.replace(prefixoCompleto, prefixoAbreviado)
                break
        return interface
    
    return abrevia(row['INT_A_APOIO']), abrevia(row['INT_B_APOIO'])

# Tratar Juniper
# Tratar Nokia
def interfacesJuniperNokia(row):
    interfaceA = row['INT_A_APOIO']
    interfaceB = row['INT_B_APOIO']
    
    if (row['VENDOR_A'].lower() == 'juniper' and row['VENDOR_B'].lower() == 'nokia') or (row['VENDOR_A'].lower() == 'nokia' and row['VENDOR_B'].lower() == 'juniper'):
        if 'rcgn' not in row['HOST_A'] or 'rcgn' not in row['HOST_B']:
            if row['VENDOR_A'].lower() == 'nokia':
                interfaceA, interfaceB = interfacesJuniperENokia('A', row)
            if row['VENDOR_B'].lower() == 'nokia':
                interfaceA, interfaceB = interfacesJuniperENokia('B', row)
    else:
        # JUNIPER
        if row['VENDOR_A'].lower() == 'juniper' and '.' not in interfaceA:
            interfaceA, interfaceB = interfacesJuniper('A', row)
        if row['VENDOR_B'].lower() == 'juniper' and '.' not in interfaceB:
            interfaceA, interfaceB = interfacesJuniper('B', row)
        # NOKIA 
        #if row['VENDOR_A'].lower() == 'nokia' and '/' in interfaceA:
            #interfaceA, interfaceB = interfacesNokia('A', row)
        #if row['VENDOR_B'].lower() == 'nokia' and '/'  in interfaceB:
            #interfaceA, interfaceB = interfacesNokia('B', row)
    
    return interfaceA, interfaceB


# Tratar Juniper x Nokia
def interfacesJuniperENokia(ponta, row):
    interfaceA = row['INT_A_APOIO']
    interfaceB = row['INT_B_APOIO']
    
    if ponta == 'A':
        operatorB = (rsnac_df['INT_B'] == row['INT_B'])
        if '.' in row['INT_B']:
            operatorA = (rsnac_df['INT_A'].str.contains('-'))
        else:
            operatorA = (rsnac_df['INT_A'].str.contains('/'))
    else:
        operatorA = (rsnac_df['INT_A'] == row['INT_A'])
        if '.' in row['INT_A']:
            operatorB = (rsnac_df['INT_B'].str.contains('-'))
        else:
            operatorB = (rsnac_df['INT_B'].str.contains('/'))
    
    df = rsnac_df[(rsnac_df['HOST_A'] == row['HOST_A']) & 
             (rsnac_df['HOST_B'] == row['HOST_B']) &
             operatorA &
             operatorB]
    
    if df.shape[0] > 0:        
        for index, row in df.iterrows():
            if ponta == 'A':
                interfaceA = df.loc[index, 'INT_A']
            else:
                interfaceB = df.loc[index, 'INT_B']
    
    return interfaceA, interfaceB
    
    
def interfacesJuniper(ponta, row):
    interfaceA = row['INT_A_APOIO']
    interfaceB = row['INT_B_APOIO']
    
    if ponta == 'A':
        operatorA = (rsnac_df['INT_A'].str.startswith(row['INT_A']+'.'))
        operatorB = (rsnac_df['INT_B'] == row['INT_B'])
    else:
        operatorA = (rsnac_df['INT_A'] == row['INT_A'])
        operatorB = (rsnac_df['INT_B'].str.startswith(row['INT_B']+'.'))
        
    df = rsnac_df[(rsnac_df['HOST_A'] == row['HOST_A']) & 
             (rsnac_df['HOST_B'] == row['HOST_B']) &
             operatorA &
             operatorB]
    
    if df.shape[0] > 0:        
        for index, row in df.iterrows():
            if ponta == 'A':
                interfaceA = df.loc[index, 'INT_A']
            else:
                interfaceB = df.loc[index, 'INT_B']
    
    return interfaceA.replace('-', ''), interfaceB.replace('-', '')


def interfacesNokia(ponta, row):
    interfaceA = row['INT_A_APOIO']
    interfaceB = row['INT_B_APOIO']
    
    if ponta == 'A':
        operatorA = (rsnac_df['INT_A'] != row['INT_A'])
        operatorB = (rsnac_df['INT_B'] == row['INT_B'])
    else:
        operatorA = (rsnac_df['INT_A'] == row['INT_A'])
        operatorB = (rsnac_df['INT_B'] != row['INT_B'])
        
    df = rsnac_df[(rsnac_df['HOST_A'] == row['HOST_A']) & 
             (rsnac_df['HOST_B'] == row['HOST_B']) &
             operatorA &
             operatorB]
    
    if df.shape[0] > 0:        
        for index, row in df.iterrows():
            if ponta == 'A':
                interfaceA = df.loc[index, 'INT_A']
            else:
                interfaceB = df.loc[index, 'INT_B']
    
    return interfaceA, interfaceB

# Tratar Nokia x Cisco e Nokia x Huawei
def interfacesNokiaECiscoEHuawei(row, vendor):
    interfaceA = row['INT_A_APOIO']
    interfaceB = row['INT_B_APOIO']
    ponta = ''
    
    if (row['VENDOR_A'].lower() == 'nokia' and row['VENDOR_B'].lower() == str(vendor).lower()) or (row['VENDOR_A'].lower() == str(vendor).lower() and row['VENDOR_B'].lower() == 'nokia'):    
        # ponta A
        if (row['VENDOR_A'].lower() == 'nokia' and row['VENDOR_B'].lower() == str(vendor).lower()):
            ponta = 'A'
            operatorA = (rsnac_df['INT_B'] == str(row['INT_B']))
            operatorB = (~rsnac_df['INT_A'].str.contains('/'))

        # ponta B
        if (row['VENDOR_A'].lower() == str(vendor).lower() and row['VENDOR_B'].lower() == 'nokia'):
            ponta = 'B'
            operatorA = (rsnac_df['INT_A'] == str(row['INT_A']))
            operatorB = (~rsnac_df['INT_B'].str.contains('/'))

        df = rsnac_df[(rsnac_df['HOST_A'] == str(row['HOST_A'])) & 
                 (rsnac_df['HOST_B'] == str(row['HOST_B'])) &
                 operatorA &
                 operatorB]

        if df.shape[0] > 0:        
            for index, row in df.iterrows():
                if ponta == 'A':
                    interfaceA = df.loc[index, 'INT_A']
                else:
                    interfaceB = df.loc[index, 'INT_B']
      
    return interfaceA, interfaceB

# Corrigir Nokia x Nokia
def interfacesNokiaNokia(row):
    interfaceA = row['INT_A_APOIO']
    interfaceB = row['INT_B_APOIO']
    
    if (row['VENDOR_A'].lower() == 'nokia' and row['VENDOR_B'].lower() == 'nokia') or (row['VENDOR_A'].lower() == 'nokia' and row['VENDOR_B'].lower() == 'nokia'):    
        interfaceA = row['INT_A']
        interfaceB = row['INT_B']
      
    return interfaceA, interfaceB
```

```python
# Campo: PadraoLabelDoc
# Informação: DESCR_A / DESCR_B
# Tipo: 
    # Se SP_FSP = FSP então EILD = NULO
    # Modelo: 
        # Legenda:
            # 3crc = 3 caracteres (letras ou _)
            # 4num = 4 números (0 a 9)
        # Padrão:
            # 3crc + _ + 3crc + - + 3crc + _ + 3crc + - + TesteVelocidade + - + 4num
            # CAS_CL_-CAS_DL_-GBE10-0001
            # Se tem EILD (FSP) - EILD
            # Se SiteA=siteB (Cidade+Site) - LOCAL
# Passos: 
# 1 - TIPO CONEXÃO
def defineIP(hostA, hostB):
    #descA = hostA.split('-')[-2].upper()
    #descB = hostB.split('-')[-2].upper()
    descA = hostA.upper()
    descB = hostB.upper()
    tecnA = ''
    tecnB = ''
    
    for i in arrayCamada:
        if descA==i[0] and descB==i[1]:
            tecnA = 'IP_'+i[2]
            tecnB = 'IP_'+i[3]
            break

    return tecnA, tecnB, descA, descB

# 2 - TAXA VELOCIDADE
def defineVelocidade(velocidade):
    try: 
        if velocidade >= 1:
            return str(round(velocidade))+'G', str(round(velocidade))+'G'
        elif velocidade == 0:
            return '', ''
        else:
            return str(round(velocidade*1000))+'M', str(round(velocidade*1000))+'M'
    except:
        return '', ''

# 3 - HOST (HOST_B->DESCR_A - HOST_A->DESCR_B)

# 4 - PORTA REMOTA
def definePorta(intA, intB):  
    #portaA = intB.replace("-", "")
    #portaB = intA.replace("-", "")
    portaA = intB
    portaB = intA
    
    return portaA, portaB

# 5 - MEIO TX (Designador)

# 6 - CGNAT (RGCN)
def adicionaCGNAT(interface):
    switcher = {
        '.10': '#INSTANCIAS-GLOBAIS',
        '.20': '#INSTANCIAS-NAT',
        '.30': '#SERVICO-LOGGING-NAT'
    }
    #print(interface + '###' + interface[-3:] + '###' + switcher.get(interface[-3:], ''))
    return switcher.get(interface[-3:], '')
    

# GERA PADRÃO LABEL DOCUMENTAÇÃO    
def fill_PadraoLabelDoc(row):
    # 1 - TIPO CONEXÃO
    #connA, connB, camadaA, camadaB = defineIP(row['HOST_A'], row['HOST_B'])
    connA, connB, camadaA, camadaB = defineIP(row['CamadaA'], row['CamadaB'])
    # 2 - TAXA VELOCIDADE
    velA, velB = defineVelocidade(row['Velocidade'])
    # 3 - HOST (HOST_B->DESCR_A - HOST_A->DESCR_B)
    hstA, hstB = row['HOST_B'], row['HOST_A']
    # 4 - PORTA REMOTA
    ptA, ptB = definePorta(row['INT_A_APOIO'], row['INT_B_APOIO'])
    # 5 - MEIO TX (Designador)
    mtx = row['Designador']
    # 6 - CGNAT (RGCN)
    cgnat = ''
    if 'rcgn' in row['HOST_A']:
        cgnat = adicionaCGNAT(row['INT_A'])
    elif 'rcgn' in row['HOST_B']:
        cgnat = adicionaCGNAT(row['INT_B'])
    
    
    
    def verificaOK(string):
        return 'OK' if len(str(string))>0 else 'NOK'
    
    qualidadeLabelA = 'Conexão: ' + verificaOK(connA) + ' - Taxa Velocidade:' + verificaOK(velA) + ' - Host: ' + verificaOK(hstA) + ' - Porta: ' + verificaOK(ptA) + ' - Designador: ' + verificaOK(mtx)
    qualidadeLabelB = 'Conexão: ' + verificaOK(connB) + ' - Taxa Velocidade:' + verificaOK(velB) + ' - Host: ' + verificaOK(hstB) + ' - Porta: ' + verificaOK(ptB) + ' - Designador: ' + verificaOK(mtx)

    labelA = connA+'#'+velA+'#'+hstA+'#'+ptA+'#'+mtx + cgnat
    labelB = connB+'#'+velB+'#'+hstB+'#'+ptB+'#'+mtx + cgnat
    padraoA = labelA if len(connA)>0 and len(velA)>0 and len(hstA)>0 and len(ptA)>0 and len(mtx)>0 else ''
    padraoB = labelB if len(connB)>0 and len(velB)>0 and len(hstB)>0 and len(ptB)>0 and len(mtx)>0 else ''
    #padraoA = labelA
    #padraoB = labelB
    
    return padraoA, padraoB, qualidadeLabelA, qualidadeLabelB, camadaA, camadaB
```

```python
# Campo: ModeloNovo
# Regras:
    # O que  foi possível construir label, comparar com DESCR_A e DESCR_B
# Passos: 
    # 1 - DESCR_A / DESCR_B = PadraoLabelDocA / PadraoLabelDocB
    # 2 - Modelo Novo = DESCR_A e DESCR_B
def fill_Modelo(row):
    ModeloNovo = False
    DescrAfull = row['DESCR_A']
    DescrBfull = row['DESCR_B']
    # Verificar observações inseridas pela operação
    DescrAsplit = DescrAfull.split('#')
    DescrBsplit = DescrBfull.split('#')

    # Verificar CGNAT
    arrayCGNAT = ['INSTANCIAS-GLOBAIS', 'INSTANCIAS-NAT', 'SERVICO-LOGGING-NAT']
    try:
        lenSplitA = 7 if str(DescrAsplit[-1]) in arrayCGNAT or str(DescrAsplit[-2]) in arrayCGNAT else 6
    except:
        lenSplitA = 6
        
    try:
        lenSplitB = 7 if str(DescrBsplit[-1]) in arrayCGNAT or str(DescrBsplit[-2]) in arrayCGNAT else 6
    except:
        lenSplitB = 6
    
    DescrA =  DescrAfull.replace('#'+str(DescrAsplit[-1]), '') if len(DescrAsplit)==lenSplitA else DescrAfull
    DescrB =  DescrBfull.replace('#'+str(DescrBsplit[-1]), '') if len(DescrBsplit)==lenSplitB else DescrBfull
    
    if DescrA.upper()==row['PadraoLabelDocA'].upper() and DescrB.upper()==row['PadraoLabelDocB'].upper():
        ModeloNovo = True
    return ModeloNovo
```

```python
# Bundle
def limpaInt(descr, host):
    interface = descr
    if '#' in interface:
        interface = interface.split('#')[3]
    else:
        interface = interface.replace('#', ' ').replace(host, '').replace('LOCAL', '')
        interface = interface.replace(interface[:interface.find(' ')], '')
        interface = interface.replace(' ', '')
    return interface

def bundle(row, intAbundle, hostAbundle, intBbundle, hostBbundle, interfacesBundle):    
    intAApoio = row['INT_A_APOIO']
    intBApoio = row['INT_B_APOIO']
    bundle = row['Bundle']
            
    if (str(row['HOST_A']) == str(hostAbundle) and str(row['HOST_B']) == str(hostBbundle)) and (str(row['INT_A']) in interfacesBundle or str(row['INT_B']) in interfacesBundle):
        if bundle == 'N':
            bundle = 'I'
            intBApoio = limpaInt(str(row['DESCR_A']), str(hostBbundle))
            intAApoio = limpaInt(str(row['DESCR_B']), str(hostAbundle))

    return intAApoio, intBApoio, bundle
```
